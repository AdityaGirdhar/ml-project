```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.array(['Paytm+Cash', 'Coupons', 'Total'])
y1 = np.array([(11.55, 17.58), (54.12, 57.99), (59.16, 60.42)])
y2 = np.array([(0.908, 0.786), (0.655, 0.604), (0.744, 0.732)])

y1m = np.array([(11.80, 18.40), (52.25, 59.81), (53.82, 59.86)])
y2m = np.array([(0.903, 0.765), (0.679, 0.579), (0.788, 0.737)])
```

```python
# plot 1
# 1 is rmse 2 is r2
# first value of tuple is without one hot, second value is with one hot
# plot r2 in second plot beside rmse

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title('RMSE vs Targets')
ax1.set_ylabel('RMSE')
ax1.set_xlabel('Targets')

ax2.set_title('R2 vs Targets')
ax2.set_ylabel('R2')
ax2.set_xlabel('Targets')

ax1.plot(x, y1[:, 0], label='Without One Hot')
ax1.plot(x, y1[:, 1], label='With One Hot')

ax2.plot(x, y2[:, 0], label='Without One Hot')
ax2.plot(x, y2[:, 1], label='With One Hot')

ax1.legend()
ax2.legend()

plt.show()
```

```python
# Plot 2
# 1 is rmse 2 is r2
# first value of tuple is without one hot, second value is with one hot
# plot r2 in second plot beside rmse

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
ax1.set_title('RMSE vs Targets (with menu)')
ax1.set_ylabel('RMSE')
ax1.set_xlabel('Targets')

ax2.set_title('R2 vs Targets (with menu)')
ax2.set_ylabel('R2')
ax2.set_xlabel('Targets')

ax1.plot(x, y1m[:, 0], label='Without One Hot')
ax1.plot(x, y1m[:, 1], label='With One Hot')

ax2.plot(x, y2m[:, 0], label='Without One Hot')
ax2.plot(x, y2m[:, 1], label='With One Hot')

ax1.legend()
ax2.legend()

plt.show()
```
