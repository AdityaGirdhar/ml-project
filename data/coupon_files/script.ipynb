```python
import pandas as pd

# Read the original CSV file
df = pd.read_csv('C:\\Users\\hp\\Desktop\\ml-practice\\Sep 23 - Sheet 1.csv')

# Initialize dictionaries to store counts for each amount
amount_counts = {2000: 0, 2500: 0, 2950: 0, 3350 : 0}

# Create a list of all dates in the month from the original data
# min_date = min(02-04-2023)
# max_date = max(df['Date'])
date_range = pd.date_range(start='09-01-2023', end='09-30-2023', freq='D')
# Format the dates in 'dd mm yyyy' format and store them in a list
formatted_dates = [date.strftime('%Y-%d-%m') for date in date_range]

# Print the formatted dates

# Create lists to store the data for the new DataFrame
count_2000 = []
count_2500 = []
count_2950 = []
count_3350 = []

print(formatted_dates)
```

```python
# Sort the DataFrame by the date column
# df['Payment Date'] = pd.to_datetime(df['Payment Date'], format='%Y-%m-%d', errors='coerce')
df['Payment Date'] = pd.to_datetime(df['Payment Date'], errors='coerce')

# df['Payment Date'] = pd.to_datetime(df['Payment Date'])
df = df.sort_values(by='Payment Date')
df = df.dropna(subset=['Payment Date'])
# print(df.info())
```

```python
# Loop through the date range to calculate counts
dates = []
count_2000 = []
count_2500 = []
count_2950 = []
count_3350 = []
for date in formatted_dates:
    # Convert date to string for comparison
    # Update counts based on the amount
    
    amount_counts = {2000: 0, 2500: 0, 2950: 0, 3350 : 0}
    for index, row in df.iterrows():
        row_date = row['Payment Date']
        # if pd.isna(row_date):
        #     print(row)
        #     break
        row_date = row_date.strftime('%Y-%m-%d')
        # print(row_date, date)
        amount = int(row['Total Amount'])
        # print(amount, type(amount))
        # print(row_date <= date and amount == 2000)
        if row_date <= date and amount == 2000: 
            amount_counts[2000] += 1
        if row_date <= date and amount == 2500:
            amount_counts[2500] += 1
        if row_date <= date and amount == 2950:
            amount_counts[2950] += 1
        if row_date <= date and amount == 3350:
            amount_counts[3350] += 1

    # print(date, amount_counts[2000])
    # print(date, amount_counts[2500])
    print(date, amount_counts[3350])
    # # Append data to the lists
    dates.append(date)
    count_2000.append(amount_counts[2000])
    count_2500.append(amount_counts[2500])
    count_2950.append(amount_counts[2950])
    count_3350.append(amount_counts[3350])

# Create a new DataFrame
# new_df = pd.DataFrame({
#     'Date': dates,
#     '15_coupon_count': count_2000,
#     '20_coupon_count': count_2500,
#     '25_coupon_count': count_2950,
#     '30_coupon_count': count_3350
    
# })


```

```python
new_df = pd.DataFrame({
    'Date': dates,
    '15_coupon_count': count_2000,
    '20_coupon_count': count_2500,
    '25_coupon_count': count_2950,
    '30_coupon_count': count_3350
    
})
```
