```python
import pandas as pd
```

```python
all_file = pd.read_csv("/Users/nalishjain/Documents/GitHub/ml-project/data/all_with_onehot.csv")
all_file.head()
```

```python
coupon_list = [
    "Oct_22_Coupons.csv",
    "Nov_22_Coupons.csv",
    "Dec_22_Coupons.csv",
    "Jan_23_Coupons.csv",
    "Feb_23_Coupons.csv",
    "Mar_23_Coupons.csv",
    "Apr_23_Coupons.csv",
    "May_23_Coupons.csv",
    "Jun_23_Coupons.csv",
    "Jul_23_Coupons.csv",
    "Aug_23_Coupons.csv",
    "Sep_23_Coupons.csv"
]

coupon_file = pd.read_csv(coupon_list[0])
coupon_file['Date'] = pd.to_datetime( coupon_file['Date'], format='%Y-%d-%m')
coupon_file['Date'] = coupon_file['Date'].dt.strftime('%Y-%m-%d')

for i in range(1, len(coupon_list)):
    new_file = pd.read_csv(coupon_list[i])
    new_file['Date'] = pd.to_datetime( new_file['Date'], format='%Y-%d-%m')
    new_file['Date'] = new_file['Date'].dt.strftime('%Y-%m-%d')
    coupon_file = pd.concat([coupon_file, new_file], ignore_index=True)

coupon_file.info()   
```

```python

# Merge DataFrames based on the 'date' column
result_df = pd.merge(all_file, coupon_file, on='Date', how='left')
result_df = result_df.dropna()

result_df.info()
```
